### Automatic propositional analysis

```python
import re
import numpy as np
from svo_parsing.subject_extraction import clean_document, extract_subjects, extract_verbs, tag_sentences, get_svos
```

```python
def load_responses(filename):
    network = [];
    with open(filename) as file:
        level = []
        for line in file:
            cleaned = re.sub('^P[0-9]+: ', '', line).strip()
            if len(cleaned) > 0:
                level.append(cleaned)
                if len(level) == 2:
                    network.append(level)
                    level = []
    return network
```

```python
chain_1 = load_responses('transcriptions/chain1.txt')
chain_2 = load_responses('transcriptions/chain2.txt')
# chain_3 = load_responses('transcriptions/chain3.txt')
# chain_4 = load_responses('transcriptions/chain4.txt')

network_1 = load_responses('transcriptions/network1.txt')
# network_2 = load_responses('transcriptions/network2.txt')
```

```python
def remove_uhms(response):
    uhm_variants = r'(ah|uhm|um|umm|uh|mm|ehm|eh|yeah|okay|ok|so|again|hey|Ah|Uhm|Um|Umm|Mm|Ehm|Eh|Yeah|Okay|Ok|So|Again|Hey)\s'
    response = re.sub(uhm_variants, '', response)
    
    uhm_variants = r'\s(ah|uhm|um|umm|uh|mm|eh|ehm|yeah|okay|okk|so|again|hey|Ah|Uhm|Um|Umm|Mm|Ehm|Eh|Yeah|Okay|Ok|So|Again|Hey)'
    response = re.sub(uhm_variants, '', response)
    
    parentheticals = r'(\(.*\)|â€¦|\.\.\.)'
    response = re.sub(parentheticals, ' ', response)
    
    multiple_whitespace = r'\s+'
    return re.sub(multiple_whitespace, ' ', response)
```

```python
import nltk
from nltk.corpus import stopwords

def remove_stopwords(response):
    return ' '.join([w for w in response.split(' ') if not w.lower() in set(stopwords.words('english'))])

from nltk.stem import WordNetLemmatizer
wnl = WordNetLemmatizer()
```

```python
chain_propositions = []
for rep, single_chain_responses in enumerate([chain_1, chain_2]): 
    replication_propositions = []
    for i, [response_1, response_2] in enumerate(single_chain_responses):
        # Collect proposiitons from response #1
        cleaned_response = clean_document(response_1)
        cleaned_response = remove_uhms(cleaned_response)
        # cleaned_response = remove_stopwords(cleaned_response)
        subjects = extract_subjects(cleaned_response)
        
        propositions = []
        for subject in subjects: # For each subject ...
            tagged_sents = tag_sentences(subject, cleaned_response)
            for sentence in tagged_sents: # For each sentence ...
                svos = get_svos(sentence, subject)
                if len(svos) > 0:
                    for svo in svos:
                        propositions.append('-'.join([svo['subject'], svo['action'], svo['object']]))
        
        # Collect propositions from response #2
        cleaned_response = clean_document(response_2)
        cleaned_response = remove_uhms(cleaned_response)
        # cleaned_response = remove_stopwords(cleaned_response)
        subjects = extract_subjects(cleaned_response)
        
        for subject in subjects:
            tagged_sents = tag_sentences(subject, cleaned_response)
            for sentence in tagged_sents:
                svos = get_svos(sentence, subject)
                if len(svos) > 0:
                    for svo in svos:
                        propositions.append('-'.join([svo['subject'], svo['action'], svo['object']]))
        
        # Append unique propositions (some may be repeated across responses)
        replication_propositions.append(set(propositions)) 
    chain_propositions.append(replication_propositions)
```

```python
for rep, single_network_responses in enumerate([network_1]): 
    network_propositions = []
    for level in single_network_responses:
        replication_propositions = []
        for i, response in enumerate(level):
            cleaned_response = clean_document(response)
            cleaned_response = remove_uhms(cleaned_response)
            # cleaned_response = remove_stopwords(cleaned_response)
            subjects = extract_subjects(cleaned_response)

            propositions = []
            for subject in subjects:
                tagged_sents = tag_sentences(subject, cleaned_response)
                for sentence in tagged_sents:
                    svos = get_svos(sentence, subject)
                    if len(svos) > 0:
                        for svo in svos:
                            propositions.append('-'.join([svo['subject'], svo['action'], svo['object']]))

            replication_propositions.append(set(propositions))
        network_propositions.append(replication_propositions)
```

### What about number of unique subjects & verbs?

```python
chain_subjects = []; chain_verbs = [];
for rep, single_chain_responses in enumerate([chain_1, chain_2]): 
    replication_subjects = []; replication_verbs = [];
    for i, [response_1, response_2] in enumerate(single_chain_responses):
        cleaned_response = clean_document(response_1)
        cleaned_response = remove_uhms(cleaned_response)
        subjects = extract_subjects(cleaned_response)
        verbs = extract_verbs(cleaned_response)
       
        cleaned_response = clean_document(response_2)
        cleaned_response = remove_uhms(cleaned_response)
        subjects += extract_subjects(cleaned_response)
        verbs += extract_verbs(cleaned_response)
        
        replication_subjects.append(set(subjects)) 
        replication_verbs.append(set(verbs)) 
    chain_subjects.append(replication_subjects)
    chain_verbs.append(replication_verbs)
```

```python
for rep, single_network_responses in enumerate([network_1]): 
    network_subjects = []; network_verbs = [];
    for level in single_network_responses:
        replication_subjects = []; replication_verbs = [];
        for i, response in enumerate(level):
            cleaned_response = clean_document(response)
            cleaned_response = remove_uhms(cleaned_response)
            subjects = extract_subjects(cleaned_response)
            verbs = extract_verbs(cleaned_response)

            replication_subjects.append(set(subjects))
            replication_verbs.append(set(verbs))
        network_subjects.append(replication_subjects)
        network_verbs.append(replication_verbs)
```
